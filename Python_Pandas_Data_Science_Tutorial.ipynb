**Loading Data into Pandas**

```python
import pandas as pd
df = pd.read_csv('pokemon_data.csv')
df.head()
```

**Reading Data in Pandas**

```python
#read headers
print(df.columns)
```

```python
#read specific datapoints, like first 5 names

print(df['Name'][0:5])
```

```python
#read specific rows pt 1

print(df.iloc[1:4])
```

```python
#read specific rows pt 2

df.loc[df['Attack'] > 150]
```

**Describing/Sorting Data**

```python
#get basic statis of the columns/variables
df.describe()
```

**Changing the Data**

create a new variable, total (a total of key stats)

```python
df['Total'] = df['HP'] + df['Attack']

df.head(5)
```

```python
#dropping some columns

df = df.drop(columns=['Total'])

df.head(5)
```
